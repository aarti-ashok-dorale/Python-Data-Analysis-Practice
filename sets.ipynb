## Sets: Mutable, no duplicates, unordered, no indexing(MU noDI)

```python
student = {101, "aarti dorale", "DS"}
print(student)
```

```python
student.add(94.60)
print(student)
```

```python
student.pop()
print(student)
```

```python
student.remove(101)
print(student) 
```

```python
# student.remove(101)
# print(student)
```

```python
student.discard(101)
print(student)
```

```python
#set operations

set1 = {1,2,3,4,5,6}
set2 = {5,6,7,8,9,10}
```

```python
print(set1.intersection(set2))
```

```python
print(set1 & set2)
```

```python
print(set1.union(set2))
```

```python
print(set1 | set2)
```

```python
print(set1.difference(set2))
```

```python
print(set2.difference(set1))
```

```python
print(set1.symmetric_difference(set2))
```

```python
print(set1 ^ set2) #cap 
```

```python
print(set1 - set2)
print(set2 - set1)
```

# Frozenset :
#Immutable version of set
#Cannot add/remove items
#Used when set must remain fixed

```python
fs = frozenset([1,2,3,4])
print(fs)
```
