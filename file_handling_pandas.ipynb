# merge 

```python
emp = {
    'id' : [1,2,3,4,5],
    'name' : ['A','B','C','D','E']   
}
df1 = pd.DataFrame(emp)
df1

emp_info = {
    'id' : [1,2,3,4,6],
    'department' : ['IT', 'IT', 'Sales', 'Sales', 'Finance']
}
df2 = pd.DataFrame(emp_info)
df2
```

```python
emp_information = pd.merge(df1,df2, on='id')
```

```python
emp_information
```

```python
left = pd.merge(df1,df2, how='left')
left
```

```python
right = pd.merge(df1,df2, how='right')
right
```

```python
outer = pd.merge(df1,df2, how='outer')
outer 
```
