```python
#17.create caculator using function 
def calculator():
    a = float(input("Enter first number: "))
    oper = input("Enter operator (+, -, *, /): ")
    b = float(input("Enter second number: "))

    if oper == '+':
        print(a + b)
    elif oper == '-':
        print(a - b)
    elif oper == '*':
        print(a * b)
    elif oper == '/':
        print(a / b)
    else:
        print("Invalid operator")

calculator() 
```

```python
#15.sort a list without using sort function

list4 = [1,2,3,8,5,4,3]

for i in range (len(list4)):
    for j in range (len(list4)):
        if list4[i] > list4[j]:
            list4[i],list4[j] = list4[j],list4[i]
print(list4) 
```

```python
#7.check if a number is palindrome in a loop
num = int(input("Enter number : "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

if num == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#7.check if a number is palindrome in a loop(second method)

st = input("enter number:")
rev = ""

for i in st:
    rev = i+rev 

if st == rev:
    print("palindrome")
else:
    print("not palindrome")
```
