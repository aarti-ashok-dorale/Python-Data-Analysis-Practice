```python
#tuple unpacking

student = (101, "aarti dorale", "DS")
print(student)
```

```python
roll_no = student[0]
name = student[1]
course_name = student[2] 
print(roll_no,name,course_name)
```

```python
roll_no,name,course_name=student
print(roll_no,name,course_name)
```
