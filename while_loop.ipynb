## while loop 

```python
i=0
while i<=10:
    print(i,end=" ")
    i+=1
```

```python
i=10
while i>=1:
    print(i,end=" ")
    i-=1
```

```python
while True:
    username=input("Enter username")
    password=input("Enter password")
    if username=="aarti" and password == "aarti123":
       print("Login Successfully")
       break
    else:
       print("Try Again") 
```

```python
#print upto 25
for i in range(1,100):
    if i == 26:
        break
    print(i,end=" ")
```

```python
# # using while loop
i = 1
while True:
    if i == 26:
        break
    print(i,end=" ")
    i += 1  
```
