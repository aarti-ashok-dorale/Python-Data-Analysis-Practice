```python
#2.even or odd

num1 = float(input("Enter a number"))

if num1%2==0:
    print("Number is even")
else:
    print("Number is odd")
```

```python
#3.pass or fail

mark = float(input("Enter marks:"))

if mark>=35:
    print("CONGRATULATIONS!!! You are passed the exam")
else:
    print("You are fail")
```

```python
#4.leap year

year = float(input("Enter year"))

if year%4==0:
    print(year,"year is leap year")
else:
    print(year,"year is not leap year") 
```

```python
#compare 3 numbers:

n1 = float(input("Enter number1"))
n2 = float(input("Enter number2"))
n3 = float(input("Enter number3"))

if n1>n2 and n1>n3:
    print("n1 is greater")
elif n2>n3 and n2>n1:
    print("n2 is greater")
elif n3>n1 and n3>n2:
    print("n3 is greater")
else:
    print("Invalid Input") 
```

```python
#age using elif
ages = int(input("Enter your age: "))

if ages >=0 and ages <= 12:
    print("Child")
elif ages > 12 and ages <= 19:
    print("Teenager")
elif ages > 19 and ages <= 35:
    print("Youth / Adult")
elif ages > 35 and ages <= 60:
    print("Working Professional")
else:
    print("Senior Citizen") 
```

```python
#Marking system using if-elif-else

marks = float(input("Enter Your marks"))

if marks==100 and marks>=90:
    print(marks,"Grade:A")
    
elif marks<90 and marks>=75:
    print(marks,"Grade:B")

elif marks<75 and marks>=60:
    print(marks,"Grade:C")

elif marks<60 and marks>=35:
    print(marks,"Grade:D")

else:
    print(marks,"Failed") 
    
```

```python
#compare 2 numbers
a = int(input("Enter 1st number"))
b = int(input("Enter 2nd number")) 

if a>b:
    print("a is greater number")
else: 
    print("b is greater number") 
```

```python
#calculator
print("***** CALCULATOR *****")
m = float(input("Enter 1st number:"))
opr = input("Enter operator")
n = float(input("Enter 2nd number"))

if opr=="+":
    print("Additon is ",m+n)
elif opr=="-":
    print("Substraction is", m-n)
elif opr=="*":
    print("Multiplicaton is", m*n)
elif opr=="/":
    print("Division is", m/n)
elif opr=="//":
    print("Floor division is", m//n)
elif opr=="%":
    print("Modulo is", m%n)
elif opr=="**":
    print("Power is", m**n)
else:
    print("operator not available")
print("***** CALCULATOR *****")
```
