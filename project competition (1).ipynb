```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("career1.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
median_cgpa = np.median(
    df['CGPA'].dropna()
)

df['CGPA'] = df['CGPA'].fillna(
    median_cgpa
)
df.head(30)  
```

```python
mode_skill = df[
'Programming_Skill'
].mode()[0]

df['Programming_Skill'] = df[
'Programming_Skill'
].fillna(mode_skill)
df.head(30)

```

```python
mean_communication = np.mean(
df['Communication_Skill'].dropna()
)

df['Communication_Skill'] = df[
'Communication_Skill'
].fillna(mean_communication)
df.head(30)
```

```python
df['Interest_Area'] = df[
'Interest_Area'
].fillna('Unknown')

df['Preferred_Work'] = df[
'Preferred_Work'
].fillna('Hybrid')

df['Internship_Experience'] = df[
'Internship_Experience'
].fillna('No')

df.head(30)
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum() 
```

```python
df.shape
```

```python
import pandas as pd

# all rows
pd.set_option('display.max_rows', None)

# all columns
pd.set_option('display.max_columns', None)

# full column content
pd.set_option('display.max_colwidth', None)

# Auto adjust width
pd.set_option('display.width', None)

recommended_career = [] 

for interest in df['Interest_Area']:

    if interest == 'AI':
        recommended_career.append('Data Scientist')

    elif interest == 'Data Science':
        recommended_career.append('Data Analyst')

    elif interest == 'Web Development':
        recommended_career.append('Software Engineer')

    elif interest == 'Cybersecurity':
        recommended_career.append('Cybersecurity Analyst')

    elif interest == 'Finance':
        recommended_career.append('Financial Analyst')

    elif interest == 'Marketing':
        recommended_career.append('Digital Marketing Specialist')

    elif interest == 'Design':
        recommended_career.append('UI/UX Designer')

    elif interest == 'Cloud Computing':
        recommended_career.append('Cloud Engineer')

    else:
        recommended_career.append('Career Counselling Needed')

# Create new column
df['Recommended_Career'] = recommended_career

# Show complete dataframe
df 
```

```python
df.to_csv('recommended_career_output.csv', index=False)
print("csv file saved successfully")
```
