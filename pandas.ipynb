# Data Selection Functions

```python
# 9.df['column']
# 10.df[['col1','col2']]
# 11.df.loc[2]
# 12.df.loc[:, 'name']
# 13.df.loc[2:5, ['name','age']]
# 14.df.iloc[2]
# 15.df.iloc[:, 1:3]

import pandas as pd

df = pd.DataFrame({
    "Name": ['A','B','C','D','E','F','G','H'],
    "Marks" : [50,60,76,78,90,99,87,72],
    "Age": [22,21,19,21,20,18,22,23]
})
print(df)
```

```python
# 9.df['column'] select only one column
print(df['Name'])
```

```python
# 10.df[['col1','col2']]

print(df[['Name','Age']])
```

```python
# 11.df.loc[2] : label (index) use karun row select : loc means location 

print(df.loc[2])
```

```python
# 12.df.loc[:, 'name'] 
print(df.loc[:, 'Name']) #  ":"=> all rows and specific columns 
```

```python
# 13.df.loc[2:5, ['name','age']]  : rows (2 to 5) + specific columns

print(df.loc[2:5,['Name', 'Age']])
```

```python
# 14.df.iloc[2] : position (index number) used karun row

print(df.iloc[2])
```

```python
#difference between loc and iloc :

print(df.loc[3]) #(label / name used krto)
print(df.iloc[0]) # (position / index used krto )
```

```python
# print(df.iloc[2])
# print(df.loc[2])

# differnece : loc[2] => index ch naav gheto
#            : iloc[2] => index gheto 

# eg;

import pandas as pd

df = pd.DataFrame({
    "Name": ["A","B","C"]}, index = [10,20,30]
)
print(df)

print(df.loc[20]) #name
print(df.iloc[2]) #index
```

```python
# 15.df.iloc[:, 1:3]

print(df.iloc[:, 1:3])
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Aarti", "karan", "tanvi", "shradha"],
    "Age" : [22, 23, 25, 30],
    "City": ["Pune", "Mumbai", "Pune", "Nagpur"],
    "Marks": [60, 90, 30, 78]
})
df
```

```python
[df['Age'] > 18] # this condition satisfy then true value
```

```python
# 1.df[df['Age'] > 18]
df[df['Age'] > 18] # in this's df is to printout the condition satisfy values
```

```python
# 2.df[df['City'] == 'Pune']
df[df['City'] == "Pune"]
```

```python
# 3.df[(df['Age'] > 18) & (df['City'] == 'Pune')]
df[(df['Age'] > 18) & (df['City'] == "Pune")] 
```

```python
# 4.df[df['Marks'].between(60, 90)]

df[df['Marks'].between(65, 90)]
```

```python
import pandas as pd

data = {
    'id' : [1,2,3,4,5,6,7],
    'name' : ['aarti','priya','shyam','ram','meean','reena',None],
    'department' : ['IT','IT','Finance','Analyst','IT',None,None],
    'Salary': [20000,30000,40000,50000,60000,60000, None]
    
}
df = pd.DataFrame(data)
df
```

```python
df.isnull().sum() 
```

```python
df.fillna({'name': 'not mentioned'}, inplace = True)
```

```python
df.loc[5,'department'] = 'Manager'
df
```

```python
df.loc[6,'department'] = 'IT'
df
```

```python
df.fillna(df['Salary'].mean(), inplace = True)
df
```

```python
df.loc[len(df)] = [8, 'mahesh', 'HR', 90000]
df
```
