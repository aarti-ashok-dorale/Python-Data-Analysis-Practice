```python
i=1
while i<=10:
    if i==5 or i==6 or i==9:
        i+=1
        continue
    print(i)
    i+=1 
```

```python
# using while loop mixture of break and continue:
i=1
while i<=100:
    if i==4 or i==8:
        i+=1
        continue
    if i==26:
        break
    print(i,end=" ")
    i+=1 
        
```
