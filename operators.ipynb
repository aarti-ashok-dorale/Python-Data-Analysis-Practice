# operators: 
1.arithematic => +,-,*,/,//,**,%
2.comparison => <,>,<=,>=,
3.assignment => +=,-=, *=, /=
4.logical => and, or, not
and => both condition be true then o/p be true
or => only one condition true then o/p be true
not => reverse the given condition
5.membership => in, not in
6.identity => is, is not
7.bitwise => &, |, ^, ~


```python
a = 10
b = 20
print(a+b)
```

```python
# Area of square : side*side
side = int(input("Enter value:"))
print(side * side)
```

```python
# Area of rectangle: length*breadth
length = int(input("Enter value of length:"))
breadth = int(input("Enter value of breadth:"))
print(length * breadth) 
```

```python
# area of circle : 3.14 * r * r
r = float(input("Enter value of radius"))
print(3.14*r*r)
```

```python
# # Area of triangle :1/2(0.5)* base*height
base = int(input("Enter value of base:"))
height = int(input("Enter value of height:"))
print(0.5 * base * height) 
```
