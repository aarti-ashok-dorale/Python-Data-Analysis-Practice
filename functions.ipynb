```python
#recursive
def fact(n):
    if n==1 or n==0:
        return n
    else:
       return n*fact(n-1)
        
fact(6)
```

```python
def naturalSum(n):
    if n==0 or n==1:
        return n
    else:
        return n+naturalSum(n-1)
naturalSum(6)

```

```python
def fibo(n):
    if n==0 or n==1:
        return n
    else:
        return fibo(n-1) + fibo(n-2)


print(fibo(6)) 
```

```python
def fibo_series(n, a=0, b=1):
    if n == 0:
        return
    print(a, end=" ")
    fibo_series(n-1, b, a+b)

fibo_series(6)  #0 index 0
                # 1 index 1
                # 1 index 2
                # 2 index 3
                # 3 index 4
                # 5 index 5 
```
