# Loops: iteration, repetition
#1.for loop -- use when, if we know range
#2.while loop -- use when, if we know the range and as well as we also don't know the range (in both condition)

## For loop

```python
for i in range(10,21,1):
    print(i,end=" ")
```

```python
for i in range(10,21):
    print(i,end=" ") # bydefault difference is 1
```

```python
for i in range(11):
    print(i,end=" ") #bydefault start with 0
```

```python
#table 

table = int(input("Enter number for table")) 

for i in range(1,11):
    print(table,"*",i,"=",table*i) 
```

```python
#even

for i in range(40):
    if i%2==0:
        print(i,end=" ") 
```

```python
#odd

for i in range(40):
    if i%2!=0:
        print(i,end=" ") 
```

```python
#sum of natural numbers:
sum = 0
for i in range(1,20):
    sum+=i
print(sum)
 
```

```python
#print +ve or -ve numbers from list
num = [2,5,76,87654,5849,654,-6,56,-78,23232,68,-5,-8]
for i in num:
    if i<0:
        print(i) 
```

```python
#endswith, startswith
fruits = ["mango","banana","pineapple","pomogranate","kiwi","papaya"]
for i in fruits:
    if i.endswith("e"):               #if i.endswith("e")
        print(i) 
```

```python
string = "Aarti"
vowels = "aeiouAEIOU"

v=0
c=0

for i in string:
    if i in vowels:
        v+=1

    else: 
        c+=1

print(v)
print(c)    
```

```python
#filtering even and odd numbers in the form of list:

list1 = [1,2,67,893,4,53,6,78,81,96,10,44,98,34]
even = []
odd = []
for i in list1:
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)
print(even)
print(odd)
        
```
