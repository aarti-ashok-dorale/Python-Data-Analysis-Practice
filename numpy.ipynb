# Broadcasting practice :- 

```python
#create arrays of different shapes,
#perform operations using braodcasting
# add a row vector to each row of matrix 
#1D
import numpy as np

array1 = np.array([1,2,3])
array2 = np.array([4,5,6]) 

addition = array1 + array2

print(addition)

vector = np.array([10,20,30])

result = addition + vector

print("result")
```

```python
#2D
array3 = np.array([[1,2,3],[4,5,6]])

vector = np.array([[10,20,30],[40,50,60]])

result = array3 + vector

print(result)
```

# Grade Analyzer :

```python
# input student marks
# output 
#1.grades :a/b/c based on average 
#2.class average 
#3.top performer

marks = np.array([20,30,60,50,46,80,91])

#2.Average
average = np.mean(marks)
print(average)

#3. Top performer
topper = np.max(marks)
print(topper)

#1.grades
grades = np.where((marks>=35) & (marks<=45),"C", np.where((marks>45) & (marks<=60),"B", np.where((marks>60) & (marks<=85),"A", np.where((marks>90),"A+","Fail")))) 
print(grades)
```
