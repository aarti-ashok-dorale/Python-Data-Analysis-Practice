```python
subject=["DS","AI","cloud","java"] #list
marks=[50,60,70,90]
sub=zip(subject,marks) #zip() function for coverting list into dictionary
sub=dict(sub)
print(sub) 
# print(sub) <zip object at 0x00000147CCB9AA40> : zip object jo process madhey aahe. 
```

```python
total_marks = sum(sub.values())
print(total_marks) 
```

```python
per=(total_marks/len(sub.values()))
print(per)
```

```python
sub.pop("java")
print(sub) 
```

```python
subject=list(sub.keys())
print(subject) 
```
