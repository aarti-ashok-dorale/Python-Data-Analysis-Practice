```python
class BankAccount:
    def __init__(self, accountholder, balance):
        self.accountholder = accountholder
        self.balance = balance
        print("accountholder:",accountholder,"balance:",balance)

    def deposit(self, amount):
        self.balance += amount
        print("deposit:",amount,"," "new balance:",self.balance)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("amount withdraw:",amount)
        else:
            print("Insufficient balance") 

    def display_balance(self):
        print("total balance:",self.balance)


acc = BankAccount("aarti,", 1000)

acc.deposit(500) 
acc.withdraw(200)
acc.display_balance()  
```

# 1.single inheritance 

```python
#single inheritance :  one parent class inherit another child class

class School:  #parent class 
    def __init__(self, sname, owner, since): #constructor
        self.sname = sname #instance attributes
        self.owner = owner
        self.since = since

    def show_school(self): #to print info
        print(self.sname, self.owner, self.since)


class Result(School):  #child class that inherits parent class
    def __init__(self, sname, owner, since, rollno, marks):
        super().__init__(sname, owner, since) #parent constructor calling 
        self.rollno = rollno
        self.marks = marks

    def show_result(self):
        print(self.rollno)

        per = sum(self.marks)/len(self.marks)
        if per < 35 :
          print("fail")
        else:
            print("pass") 


r = Result("puneripattern", "miss sarika mam", 2000, 10, [20,30,30,50,50]) 
```

```python
r.show_school()
```

```python
r.show_result()
```

# 2.Multilevel inheritance 

```python
# <!-- # define : A class inherits from a parent, and another class inherits from that child.
# # Q.1  -->
```

```python
class School: #parent class
    def __init__(self, sname, owner, since): #constructor
        self.sname = sname #instance attributes
        self.owner = owner
        self.since = since

    def show_school(self): #to show info of school
        print("school name:",self.sname, "\n owner name:",self.owner, "\n since:",self.since) 

class Course(School): #child class inherits parent class(school)
    def __init__(self,sname,owner,since,cname,duration,fees):
        super().__init__(sname, owner, since) # parent constructor calling
        self.cname = cname
        self.duration = duration 
        self.fees = fees

    def course_details(self): 
        print("course name:",self.cname, "\nduration:", self.duration, "\nfees:",self.fees) 


class Student(Course): #Another child class inherits parent class(course)
    def __init__(self,sname,owner,since,cname,duration,fees,stud_name,age,attendance):
        super().__init__(sname,owner,since,cname,duration,fees) # parent constructor calling
        self.stud_name = stud_name
        self.age = age
        self.attendance = attendance 

    def student_details(self):
        print("student name:", self.stud_name, "\nage:", self.age, "\nattendance : ", self.attendance)

class Result(Student):  # final child class inherits parent class(student)
    def __init__(self,sname,owner,since,cname,duration,fees,stud_name,age,attendance,rollno,marks):
        super().__init__(sname,owner,since,cname,duration,fees,stud_name,age,attendance) # parent constructor calling 
        self.rollno = rollno
        self.marks = marks 

    def show_result(self):
        print(self.rollno)

        per = sum(self.marks)/len(self.marks)
        if per < 35 :
          print("fail")
        else:
            print("pass") 


m = Result("puneri pattern","sarika mam",2020,"DS","6 mon", 60000,"aarti",22,"50%",11,[20,40,50,50,30]) #object creation
```

```python
m.show_school()
```

```python
m.course_details()
```

```python
m.student_details()
```

```python
m.show_result()
```

# 3.Multiple Inheritance

```python
#define : two or more parent class inherit only one child 

class book1:     #1st parent class
    def __init__(self,bookname1,author1,price1):
        self.bookname1 = bookname1       #instance attributes
        self.author1 = author1
        self.price1 = price1

    def show_book1(self): #To print info
        print("\nbookname1:",self.bookname1,"\nauthor1:",self.author1,"\nprice1:",self.price1) 

class book2: #2nd parent class
    def __init__(self,bookname2,author2,price2):
        self.bookname2 = bookname2
        self.author2 = author2
        self.price2 = price2

    def show_book2(self):
        print("\nbookname2:",self.bookname2,"\nauthor2:",self.author2,"\nprice2:",self.price2) 

class Libraries(book1,book2):     #child class inherits both classes
    def __init__(self,bookname1,author1,price1,bookname2,author2,price2,lname,location):
        self.lname = lname
        self.location = location
        book1.__init__(self,bookname1,author1,price1) # book1 class's constructor (__init__) manually calling
        book2.__init__(self,bookname2,author2,price2) # book1 class's constructor (__init__) manually calling

    def show_libraries(self):
        print("\nlibrary name:",self.lname,"\nlocation:",self.location) 


b = Libraries("Wings Of Fire","DR.A.P.J Abdul Kalam",300,"Collector Sahiba","MR.pratik yadav",250,"PP","Magarpatta")  #object of Libraries
```

```python
b.show_book1() #method calling
```

```python
b.show_book2()
```

```python
b.show_libraries()
```

# 4 .Hierarchical Inheritance 

```python
#definition : one parent class inherits two or more child classes

class library:
    def __init__(self,lname,location,city="Pune"):
        self.lname = lname
        self.location = location
        self.city = city

    def display_lib(self):
        print("\nlibrary name:",self.lname, "location:",self.location,"city:",self.location)

class book1(library):
    def __init__(self,lname,location,bname1,author1,price1,city="Pune"):
        super().__init__(lname,location,city="Pune")
        self.bname1 = bname1
        self.author1 = author1
        self.price1 = price1

    def show_book1(self):
        print("\nbook name:",self.bname1,"\nauthor name:",self.author1,"\nprice:",self.price1)

class book2(library):
    def __init__(self,lname,location,bname2,author2,price2,city="Pune"):
        super().__init__(lname,location,city="Pune")
        self.bname2 = bname2
        self.author2 = author2
        self.price2 = price2

    def show_book2(self):
        print("\nbook name:",self.bname2,"\nauthor name:",self.author2,"\nprice:",self.price2) 


b1 = book1("PP","Pune","wings of fire","A.P.J Abdul kalam",200)
b2 = book2("DD","Hadapsar","abc","xyz",300)        
```

# 5.Hybrid Inheritance

```python
# definition: : combintion of single,hierarchical,multiple,multilevel inheritance 

class student:
    def __init__(self,name,rollno):
        self.name = name
        self.rollno = rollno

    def display_student(self):
        print("student name:",self.name,"roll no:",self.rollno)

class academic(student):
    def __init__(self,name,rollno,marks):
        student.__init__(self,name,rollno)
        self.marks = marks

    def academic_result(self):
        print("academic marks:", self.marks) 

class sport(student):
    def __init__(self,name,rollno,s_score):
        student.__init__(self,name,rollno) 
        self.s_score = s_score

    def sport_result(self):
        print("sport score:", self.s_score) 

class result(academic,sport): #multiple
    def __init__(self,name,rollno,marks,s_score):
        academic.__init__(self,name,rollno,marks)
        sport.__init__(self,name,rollno,s_score)

    def cal_result(self):
        return sum(self.marks) + self.s_score

    def display_result(self):  #display complete result
        self.display_student()
        self.academic_result()
        self.sport_result() 
        print("Total marks :",self.cal_result())

b1 = result("aarti",101,[56,70,50,40],30)   
```

```python
b1.display_result()  
```
