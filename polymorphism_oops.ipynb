```python
#using inheritance
class Animal:
    def sound(self):
        print("Animal makes sound")

class Cat(Animal):
    def sound(self):
        super().sound()
        print("cat makes sound - mew")

c = Cat()
c.sound() 
```

```python
# with class 
class Animal:
    def sound(self):
        print("Animal makes sound")

class Cat:
    def sound(self):
        print("cat makes meww")

for i in Animal(),Cat():
    i.sound() 
```

```python
#duck method
class Animal:
    def sound(self):
        print("Animal makes sound")

class Cat:
    def sound(self):
        print("cat makes mew") 

def make_sound(a):
    a.sound()

make_sound(Animal())
make_sound(Cat())
```

```python
# generators : use yeild keyword instead of return
def gen():  #user defined func
    for i in range(1,11):
        yield i
        
for i in gen():
    print(i)
```

```python
# decorator : 

def decorator(func): #user defined func
    def wrapper(): #user defined func
        print("this is wrapper function")
        func()
        print("this is main decorator function")

    return wrapper 

@decorator
def hello():
    print("hello")

hello()
```

```python
logged_in_user = None

def login_required(func):
    def wrapper(*arg,**kwarg):
        if logged_in_user is None:
            print("Access Denied")
        else:
            return func(*arg,**kwarg) 
    return wrapper

@login_required
def view_dashboard():
    print("welcome to dashboard\n",logged_in_user,"this is your dashboard")

view_dashboard() 
```
